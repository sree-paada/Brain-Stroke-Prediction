#Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df.head()
```

```python
df.shape
```

```python
# removing neccessary duplicates 
df = df.drop_duplicates()
df.shape
```

```python
# checking missing values 
df.isnull().sum()
```

```python
#Filling the missing values
df.fillna(value = df['bmi'].median(), inplace = True)
df.isnull().sum()
```

```python
df.drop(columns = 'id', inplace = True)
d = df[df['gender'] == 'Other'].index
df.drop(d, inplace = True)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
```

```python
gender=enc.fit_transform(df['gender'])
smoking_status=enc.fit_transform(df['smoking_status'])
work_type=enc.fit_transform(df['work_type'])
Residence_type=enc.fit_transform(df['Residence_type'])
ever_married=enc.fit_transform(df['ever_married'])
```

```python
df['ever_married']=ever_married
df['Residence_type']=Residence_type
df['smoking_status']=smoking_status
df['gender']=gender
df['work_type']=work_type
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

#Data Preparation

```python
X = df.drop(['stroke'], axis=1)
y = df['stroke']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=12)
```

```python
print('X train is ',X_train.shape)
print('Y train is ',y_train.shape)
print('X test is ',X_test.shape)
print('Y test is ',y_test.shape)
```

##Smote Technique

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

counter = Counter(y_train)
print(counter)
smt = SMOTE(random_state=12)
X_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)

counter = Counter(y_train_sm)
print(counter)
```

#XGboost Random Forest

```python
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, classification_report
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train_sm, y_train_sm)
#print(xgb_model.score(X_train, y_train))

print(xgb_model.score(X_test, y_test))
```

```python
y_pred = xgb_model.predict(X_test)
pred = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
print(classification_report(y_test, pred))
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis = LogisticRegression(solver = 'liblinear')
logis.fit(X_train_sm, y_train_sm)
```

```python
train_pred = logis.predict(X_train_sm)
pred = logis.predict(X_test)
```

```python
pred = [round(value) for value in pred]
```

```python
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
print(classification_report(y_test, pred))
```

#Gaussian Naives

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train_sm,y_train_sm)
```

```python
pred = gnb.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
print(classification_report(y_test, pred))
```
